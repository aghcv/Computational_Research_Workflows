## Why visualization matters

Visualization helps researchers inspect data quality, identify trends, and communicate results.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(1)
n = 30
df = pd.DataFrame({
    "case": [f"case_{i:02d}" for i in range(n)],
    "severity": np.random.choice([50, 75, 95], size=n),
    "response": np.random.normal(loc=0, scale=1, size=n),
})
df["response"] += df["severity"] / 50
df.head()
```

```python
summary = df.groupby("severity")["response"].agg(["mean", "std", "count"])
summary
```

```python
plt.figure()
for severity, group in df.groupby("severity"):
    plt.scatter([severity] * len(group), group["response"], label=f"{severity}%")

plt.xlabel("Severity")
plt.ylabel("Response")
plt.title("Response by severity group")
plt.legend()
plt.show()
```
