## A first research-style computation

In research, we often start with data, compute a quantity of interest, and visualize the result. The example below creates a simple synthetic signal.

This is not a real experiment, but it has the same basic structure as many research workflows:

```text
create or load data -> analyze data -> visualize result -> interpret result
```

````python
import numpy as np
import matplotlib.pyplot as plt

time = np.linspace(0, 10, 200)
signal = np.sin(time) + 0.2 * np.random.randn(len(time))

plt.figure()
plt.plot(time, signal)
plt.xlabel("Time")
plt.ylabel("Signal")
plt.title("Synthetic research signal")
plt.show()
````

## Example: A first image-processing task

Biomedical research often involves images, including microscopy, CT, MRI, ultrasound, histology, and camera-based data.

Here, we will use a built-in sample image and apply a simple edge-detection filter.

````python
!pip install scikit-image
````

````python
from skimage import data, filters
import matplotlib.pyplot as plt

image = data.camera()
edges = filters.sobel(image)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap="gray")
plt.title("Original image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(edges, cmap="gray")
plt.title("Detected edges")
plt.axis("off")

plt.show()
````

## Research connection

Many biomedical image-analysis pipelines begin with steps like this:

- loading an image
- filtering noise
- detecting edges or boundaries
- separating objects from background
- measuring size, shape, intensity, or motion

In a real project, the image might come from a microscope, CT scanner, MRI scanner, ultrasound system, or video recording.

## Discussion: What makes a repository easier to use?

A beginner-friendly repository usually has:

- a clear `README`
- simple examples
- small data files or downloadable data
- installation instructions
- notebooks that can run in Colab
- comments explaining the code

A more advanced research repository may be harder to run at first because it may require:

- special data
- large files
- a GPU
- a specific operating system
- special software dependencies
- high-performance computing resources

Both types of repositories are useful. Colab helps us start with the easy parts first.
