## A first research-style computation

In research, we often start with data, compute a quantity of interest, and visualize the result. The example below creates a simple synthetic signal.

```python
import numpy as np
import matplotlib.pyplot as plt

time = np.linspace(0, 10, 200)
signal = np.sin(time) + 0.2 * np.random.randn(len(time))

plt.figure()
plt.plot(time, signal)
plt.xlabel("Time")
plt.ylabel("Signal")
plt.title("Synthetic research signal")
plt.show()
```
