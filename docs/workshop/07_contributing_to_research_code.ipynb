## From student to contributor

A contribution does not need to be large. Good first contributions include:

- improving documentation
- adding an example
- fixing a typo
- adding a small test
- improving a plot label
- simplifying installation instructions

## Contributor workflow

1. Find or create an issue
2. Fork the repository
3. Create a branch
4. Make changes
5. Run tests
6. Commit and push
7. Open a pull request
8. Respond to review comments

## Example branch names

```bash
docs/add-colab-instructions
test/add-pressure-drop-test
fix/readme-typo
example/add-visualization-demo
```

## Pull request checklist

Before opening a pull request, confirm:

- The code runs
- Tests pass
- The README or documentation is updated if needed
- The change is small and focused
- The pull request explains what changed and why
