# 2. What is a project folder?

A research software project is usually not just one notebook.

It may contain:

```text
my_project/
├── notebooks/
│   └── exploratory_analysis.ipynb
├── src/
│   └── analysis_utils.py
├── tests/
│   └── test_analysis_utils.py
├── data/
│   └── sample_data.csv
├── figures/
│   └── result_plot.png
├── README.md
├── requirements.txt
└── .github/
    └── workflows/
        └── tests.yml
```

VS Code is designed to open this whole folder and let you move between files easily.

# 4. Notebooks and scripts work together

A common beginner mistake is thinking that notebooks and scripts are competing tools.

They are best used together:

| Use notebooks for | Use scripts for |

## Demo: Put reusable code in a Python file

In Session 4, we tested a small function.

Here, we place that function inside a project file, like we would in VS Code.


````python
%%writefile session5_research_workspace/src/analysis_utils.py
def pressure_drop_resistance(pressure_drop, flow_rate):
    """Calculate resistance from pressure drop and flow rate.

    Parameters
    ----------
    pressure_drop : float
        Difference in pressure.
    flow_rate : float
        Flow rate. Must be positive.

    Returns
    -------
    float
        Resistance = pressure_drop / flow_rate.
    """
    if flow_rate <= 0:
        raise ValueError("flow_rate must be positive")
    return pressure_drop / flow_rate
````

# 6. VS Code extensions

VS Code starts as a general editor.

## Demo: Add a test file

We will test the function in `src/analysis_utils.py`.

This is the same concept from Session 4, now placed inside a project structure.


````python
%%writefile session5_research_workspace/tests/test_analysis_utils.py
import sys
from pathlib import Path

sys.path.append(str(Path(__file__).resolve().parents[1] / "src"))

import pytest
from analysis_utils import pressure_drop_resistance


def test_pressure_drop_resistance():
    assert pressure_drop_resistance(10, 2) == 5


def test_pressure_drop_resistance_rejects_zero_flow():
    with pytest.raises(ValueError):
        pressure_drop_resistance(10, 0)
````

# 9. Continuous integration from a developer workspace

In Session 4, CI meant:

```text
Push code to GitHub
        ↓
GitHub automatically runs tests
        ↓
The pull request shows pass/fail
```

VS Code helps because you can prepare the same files locally:

```text
write code → run tests → commit → push → CI runs on GitHub
```

A good rule:

> Run tests locally before pushing code to GitHub.


## Demo: Create a GitHub Actions workflow file

This file tells GitHub how to run tests automatically.

In VS Code, you would create this file in:

```text
.github/workflows/tests.yml
```


````python
%%writefile session5_research_workspace/.github/workflows/tests.yml
name: Tests

on:
  pull_request:
  push:

jobs:
  test:
    runs-on: ubuntu-latest

    steps:
      - uses: actions/checkout@v4

      - uses: actions/setup-python@v5
        with:
          python-version: "3.11"

      - run: pip install -r requirements.txt
      - run: pytest -q
````

# 10. Data visualization in VS Code

VS Code can help you inspect and visualize research data.

Common examples:

- preview a CSV file,
- open images such as `.png`,
- view plots generated by Python,
- run notebook cells that create figures,
- compare result files across different runs.

This is useful because research computing often involves both code and outputs.


## Demo: Create a small dataset and plot

In VS Code, you could view the CSV file and the saved figure from the file explorer.


````python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

project = Path("session5_research_workspace")

df = pd.DataFrame({
    "flow_rate": [1, 2, 3, 4, 5],
    "pressure_drop": [4, 8, 12, 16, 20],
})

df["resistance"] = df["pressure_drop"] / df["flow_rate"]

csv_path = project / "data" / "sample_pressure_flow_data.csv"
fig_path = project / "figures" / "pressure_drop_vs_flow_rate.png"

df.to_csv(csv_path, index=False)

plt.figure(figsize=(5, 4))
plt.plot(df["flow_rate"], df["pressure_drop"], marker="o")
plt.xlabel("Flow rate")
plt.ylabel("Pressure drop")
plt.title("Pressure drop vs. flow rate")
plt.grid(True)
plt.savefig(fig_path, dpi=150, bbox_inches="tight")
plt.show()

print(f"Saved data to: {csv_path}")
print(f"Saved figure to: {fig_path}")
````

# 11. Remote development and HPC

In Session 2, we discussed why laptops are not always enough for large computations.

HPC systems may provide:

- many CPU cores,
- GPUs,
- large memory,
- job schedulers,
- shared filesystems,
- specialized scientific software.

VS Code can connect to remote machines using SSH.

The important idea:

```text
Your laptop runs VS Code
        ↓
VS Code connects through SSH
        ↓
Your code runs on the remote server or HPC login/development node
```


## What SSH means

SSH stands for **Secure Shell**.

It lets you log into another computer from your terminal.

A typical command looks like:

```bash
ssh username@hpc.university.edu
```

After connecting, you are no longer working on your laptop filesystem.  
You are working on the remote machine.


## What Remote-SSH adds in VS Code

Without VS Code:

```text
Terminal only → type commands manually → edit files using terminal editors
```

With VS Code Remote-SSH:

```text
VS Code interface → remote files → remote terminal → remote Python → remote Git
```

This can make HPC workflows less intimidating for beginners.


# 12. HPC workflow from VS Code

A common workflow looks like this:

```text
1. Connect to HPC using Remote-SSH
2. Open your project folder on the remote system
3. Edit code in VS Code
4. Run small tests on the login or development node
5. Submit large jobs to the scheduler
6. Monitor job status
7. Visualize small outputs locally or in VS Code
8. Commit and push changes to GitHub
```

Important caution:

> On most HPC systems, you should not run heavy computations directly on the login node.  
> Heavy jobs usually need to be submitted to a scheduler.


## Example HPC commands

These commands are examples only.  
They depend on the specific HPC system.

```bash
# Connect
ssh username@hpc.university.edu

# Load software
module load python

# Create or activate an environment
conda activate myenv

# Run a small test
pytest -q

# Submit a larger job
sbatch run_simulation.sh

# Check job status
squeue -u username
```

The key idea is that VS Code can give you a familiar interface while these commands run remotely.
