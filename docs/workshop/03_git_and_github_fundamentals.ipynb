## Git vs. GitHub

**Git** is a version control system. It tracks changes to files over time.

**GitHub** is an online platform for hosting Git repositories and collaborating with others.

Important terms:

- **Repository:** a project folder tracked by Git
- **Commit:** a saved snapshot of changes
- **Branch:** a parallel version of the project
- **Issue:** a task, bug report, or discussion item
- **Pull request:** a proposed change submitted for review
