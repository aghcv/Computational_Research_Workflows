## Concept 1 — Why do we need version control?

Research code changes over time.

Without version control, students often create files like:

```text
analysis.py
analysis_final.py
analysis_final_v2.py
analysis_final_REAL_FINAL.py
```

This becomes confusing quickly.

**Version control** solves this by recording the history of a project in an organized way.

````python
# Visual 1: A project without version control vs. with version control

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 4))
ax.axis("off")

# Left side
ax.text(0.08, 0.85, "Without version control", fontsize=14, weight="bold")
files = ["analysis.py", "analysis_final.py", "analysis_final_v2.py", "analysis_final_REAL_FINAL.py"]
for i, f in enumerate(files):
    ax.text(0.08, 0.65 - i*0.12, f"📄 {f}", fontsize=12)

# Right side
ax.text(0.58, 0.85, "With Git", fontsize=14, weight="bold")
versions = ["Commit 1", "Commit 2", "Commit 3", "Commit 4"]
for i, v in enumerate(versions):
    x = 0.58 + i*0.10
    ax.text(x, 0.55, "●", fontsize=24, ha="center")
    ax.text(x, 0.43, v, fontsize=10, ha="center")
    if i < len(versions)-1:
        ax.annotate("", xy=(x+0.08, 0.57), xytext=(x+0.025, 0.57),
                    arrowprops=dict(arrowstyle="->", lw=1.5))

ax.text(0.58, 0.25, "One project folder + organized history", fontsize=12)
plt.show()
````

## Concept 2 — Git and GitHub are not the same thing

**Git** is the tool that tracks changes.

**GitHub** is a website where Git repositories can be stored, shared, reviewed, and tested.

A simple analogy:

```text
Git = the lab notebook system
GitHub = the shared online lab space
```

````python
# Visual 2: Git vs GitHub

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(9, 4))
ax.axis("off")

boxes = [
    (0.10, 0.45, "Your computer / Colab\nworking files"),
    (0.40, 0.45, "Git\ntracks history locally"),
    (0.70, 0.45, "GitHub\nshares project online")
]

for x, y, label in boxes:
    rect = plt.Rectangle((x, y), 0.2, 0.22, fill=False, linewidth=2)
    ax.add_patch(rect)
    ax.text(x+0.1, y+0.11, label, ha="center", va="center", fontsize=11)

ax.annotate("", xy=(0.40, 0.56), xytext=(0.30, 0.56), arrowprops=dict(arrowstyle="->", lw=2))
ax.annotate("", xy=(0.70, 0.56), xytext=(0.60, 0.56), arrowprops=dict(arrowstyle="->", lw=2))

ax.text(0.18, 0.25, "edit", ha="center", fontsize=11)
ax.text(0.50, 0.25, "commit", ha="center", fontsize=11)
ax.text(0.80, 0.25, "collaborate", ha="center", fontsize=11)

plt.show()
````

## Concept 5 — Git has three important file states

For beginners, the most important Git idea is this:

```text
Modified → Staged → Committed
```

Meaning:

1. **Modified:** you changed a file.
2. **Staged:** you selected the change for the next saved snapshot.
3. **Committed:** Git saved the snapshot in history.

This is the main mental model for using Git.

````python
# Visual 3: Modified, Staged, Committed

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 3.5))
ax.axis("off")

labels = [
    ("Modified", "You changed a file"),
    ("Staged", "You selected it\nfor the next snapshot"),
    ("Committed", "Git saved the snapshot\nin project history")
]

for i, (title, desc) in enumerate(labels):
    x = 0.10 + i*0.32
    rect = plt.Rectangle((x, 0.45), 0.22, 0.25, fill=False, linewidth=2)
    ax.add_patch(rect)
    ax.text(x+0.11, 0.62, title, ha="center", fontsize=13, weight="bold")
    ax.text(x+0.11, 0.51, desc, ha="center", fontsize=10)
    if i < 2:
        ax.annotate("", xy=(x+0.31, 0.575), xytext=(x+0.23, 0.575),
                    arrowprops=dict(arrowstyle="->", lw=2))

ax.text(0.345, 0.25, "git add", ha="center", fontsize=11)
ax.text(0.665, 0.25, "git commit", ha="center", fontsize=11)

plt.show()
````

## Concept 8 — `git add` stages a file

`git add` does **not** permanently save the file.

It only says:

> Put this file into the next snapshot.

Now `README.md` should appear under **Changes to be committed**.

This means it is staged.

## Concept 11 — A commit is not just a file copy

Git thinks of your project as a sequence of snapshots.

Each commit points to a version of the whole project.

````python
# Visual 4: Commit history as snapshots

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(9, 3))
ax.axis("off")

commits = ["c1\nAdd README", "c2\nEdit README", "c3\nAdd script"]
for i, c in enumerate(commits):
    x = 0.15 + i*0.28
    ax.text(x, 0.55, "●", fontsize=28, ha="center")
    ax.text(x, 0.35, c, fontsize=10, ha="center")
    if i < len(commits)-1:
        ax.annotate("", xy=(x+0.23, 0.58), xytext=(x+0.04, 0.58),
                    arrowprops=dict(arrowstyle="->", lw=2))
ax.text(0.5, 0.80, "Git history = snapshots over time", fontsize=14, ha="center", weight="bold")
plt.show()
````

## Concept 15 — Branches let you try changes safely

A **branch** is a separate line of work.

You can use branches to try an idea without changing the main version immediately.

This is essential for collaboration and pull requests.

````python
# Visual 5: Branching

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 4))
ax.axis("off")

# main line
points_main = [(0.15,0.55), (0.35,0.55), (0.55,0.55)]
for i, (x,y) in enumerate(points_main):
    ax.text(x, y, "●", fontsize=24, ha="center")
    if i < len(points_main)-1:
        ax.annotate("", xy=(points_main[i+1][0]-0.03,y+0.02), xytext=(x+0.04,y+0.02),
                    arrowprops=dict(arrowstyle="->", lw=2))

# branch
branch_points = [(0.35,0.55), (0.55,0.78), (0.75,0.78)]
ax.annotate("", xy=(0.52,0.76), xytext=(0.38,0.58), arrowprops=dict(arrowstyle="->", lw=2))
ax.text(0.55,0.78,"●",fontsize=24,ha="center")
ax.text(0.75,0.78,"●",fontsize=24,ha="center")
ax.annotate("", xy=(0.72,0.80), xytext=(0.59,0.80), arrowprops=dict(arrowstyle="->", lw=2))

ax.text(0.60, 0.45, "main branch", fontsize=12)
ax.text(0.62, 0.90, "feature branch", fontsize=12)
ax.text(0.5, 0.20, "Branches let you work on an idea separately.", fontsize=13, ha="center")

plt.show()
````

## Concept 20 — GitHub adds collaboration around Git

GitHub helps teams do things that Git alone does not fully manage:

- host repositories online
- open issues
- discuss changes
- review pull requests
- run automated tests
- track project history in a browser

````python
# Visual 6: GitHub collaboration workflow

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(11, 4))
ax.axis("off")

steps = [
    ("1", "Create branch"),
    ("2", "Make commits"),
    ("3", "Push to GitHub"),
    ("4", "Open pull request"),
    ("5", "Review + tests"),
    ("6", "Merge")
]

for i, (num, label) in enumerate(steps):
    x = 0.05 + i*0.155
    circle = plt.Circle((x+0.055, 0.60), 0.045, fill=False, linewidth=2)
    ax.add_patch(circle)
    ax.text(x+0.055, 0.60, num, ha="center", va="center", fontsize=12, weight="bold")
    ax.text(x+0.055, 0.40, label, ha="center", fontsize=10)
    if i < len(steps)-1:
        ax.annotate("", xy=(x+0.16,0.60), xytext=(x+0.105,0.60),
                    arrowprops=dict(arrowstyle="->", lw=1.8))

ax.text(0.5, 0.82, "Typical GitHub collaboration workflow", fontsize=14, ha="center", weight="bold")
plt.show()
````

## Concept 22 — Clone, pull, and push

Three important GitHub-connected commands:

```bash
git clone <url>
```
Copies a GitHub repository to your computer or Colab.

```bash
git pull
```
Downloads new changes from GitHub into your local copy.

```bash
git push
```
Uploads your committed changes to GitHub.

````python
# Visual 7: Clone, pull, and push

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(9, 4))
ax.axis("off")

# boxes
for x, label in [(0.15, "Local / Colab\nrepository"), (0.65, "GitHub\nrepository")]:
    rect = plt.Rectangle((x, 0.40), 0.22, 0.25, fill=False, linewidth=2)
    ax.add_patch(rect)
    ax.text(x+0.11, 0.525, label, ha="center", va="center", fontsize=12)

# arrows
ax.annotate("clone / pull", xy=(0.37,0.58), xytext=(0.65,0.58),
            arrowprops=dict(arrowstyle="->", lw=2), ha="center", fontsize=11)
ax.annotate("push", xy=(0.65,0.45), xytext=(0.37,0.45),
            arrowprops=dict(arrowstyle="->", lw=2), ha="center", fontsize=11)

ax.text(0.5, 0.22, "Only committed changes can be pushed.", ha="center", fontsize=12)
plt.show()
````

## Concept 23 — Pull requests prepare us for unit testing and CI

A **pull request** is a request to merge your branch into another branch.

In many research software projects, a pull request triggers automatic checks:

```text
student pushes code → GitHub sees change → tests run automatically → pass/fail result
```

This is the bridge to the next session.

````python
# Visual 8: Bridge to next session

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(11, 3.5))
ax.axis("off")

steps = ["Push code", "Pull request", "GitHub Actions", "Unit tests", "Pass / fail"]
for i, label in enumerate(steps):
    x = 0.08 + i*0.20
    rect = plt.Rectangle((x, 0.50), 0.15, 0.22, fill=False, linewidth=2)
    ax.add_patch(rect)
    ax.text(x+0.075, 0.61, label, ha="center", va="center", fontsize=10)
    if i < len(steps)-1:
        ax.annotate("", xy=(x+0.20,0.61), xytext=(x+0.15,0.61),
                    arrowprops=dict(arrowstyle="->", lw=2))

ax.text(0.5, 0.25, "Next session: how automated tests protect research code", ha="center", fontsize=13)
plt.show()
````
